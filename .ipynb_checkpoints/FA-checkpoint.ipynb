```python
#encoding=utf-8
import wave
import numpy as np
import os
import csv
import random
import pylab

threshold = 0.9 # 标定强度阈值（0~1之间）
key_file = "./Piano88/German Concert D "    # 钢琴88音所在目录
FDicCSV = "./Fcsv_final.csv"    # 钢琴88音对应csv文件
# 全部以简谱表示X(Y),X为音调，Y为有多少个音级
KeyList = [
    "6(-4)","6(-4)#","7(-4)",    # 大字一组
            "1(-3)","1(-3)#","2(-3)","2(-3)#","3(-3)","4(-3)","4(-3)#",
            "5(-3)","5(-3)#","6(-3)","6(-3)#","7(-3)" , # 大字二组  
            "1(-2)","1(-2)#","2(-2)","2(-2)#","3(-2)","4(-2)","4(-2)#",
            "5(-2)","5(-2)#","6(-2)","6(-2)#","7(-2)" , # 大字组 
            "1(-1)","1(-1)#","2(-1)","2(-1)#","3(-1)","4(-1)","4(-1)#",
            "5(-1)","5(-1)#","6(-1)","6(-1)#","7(-1)" , # 小字组 
            "1","1#","2","2#","3","4","4#",
            "5","5#","6","6#","7" , # 小字一组，中央调
            "1(+1)","1(+1)#","2(+1)","2(+1)#","3(+1)","4(+1)","4(+1)#",
            "5(+1)","5(+1)#","6(+1)","6(+1)#","7(+1)" , # 小字二组 
            "1(+2)","1(+2)#","2(+2)","2(+2)#","3(+2)","4(+2)","4(+2)#",
            "5(+2)","5(+2)#","6(+2)","6(+2)#","7(+2)" , # 小字三组 
            "1(+3)","1(+3)#","2(+3)","2(+3)#","3(+3)","4(+3)","4(+3)#",
            "5(+3)","5(+3)#","6(+3)","6(+3)#","7(+3)" , # 小字四组 
            "1(+4)" #小字五组
            ]
```

### 产生波形数据
选用音频[晚风](./night.wav)

```python
# 产生波形数据
def GenerateWavData(WavFile):
    # 打开WAV文档，文件路径根据需要做修改
    with wave.open(WavFile,'rb') as wf:
        # 读取WAV文档的参数
        nframes = wf.getnframes()
        framerate = wf.getframerate()
        width = wf.getsampwidth()
        channels = wf.getnchannels()
        # 读取完整的帧数据到str_data中，这是一个string类型的数据
        str_data = wf.readframes(nframes)
    # 将波形数据转换为数组
    wave_data = np.fromstring(str_data, dtype=np.short)
    # 将wave_data数组改为2列，行数自动匹配。在修改shape的属性时，需使得数组的总长度不变。
    wave_data.shape = -1,2
    # 将数组转置
    wave_data = wave_data.T
    return wave_data,nframes,framerate,width,channels

# 产生波形图
def GenerateWavGraph(WavFile):
    wave_data,nframes,framerate,width,channels = GenerateWavData(WavFile)
    # time 也是一个数组，与wave_data[0]或wave_data[1]配对形成系列点坐标
    time = np.arange(0,nframes)*(1.0/framerate)
    # #绘制波形图
    pylab.plot(time, wave_data[0])
    pylab.subplot(212)
    pylab.plot(time, wave_data[1], c="g")
    pylab.xlabel("time (seconds)")
    pylab.show()

WavFile = "./night.wav"
print("原始音频波形图如下：")
GenerateWavGraph(WavFile)
```

### 切割音频文件
将原始音频文件进行切割，得到若干子文件

```python
CutTimeDef = 1000 # 以500ms截断文件
CutFrameNum =0

def CutFile(wavFileName):
    # fin = open(wavFileName,'r')
    [filename, ext] = os.path.splitext(wavFileName) 
    print("CutFile File Name is ",filename)
    filedir = os.path.join('./',filename) + "_segmentation"
    print("The Cuts are stored in %s" %filedir)
    if not os.path.exists(filedir):
        os.mkdir(filedir)
    # 打开wav文件 ，open返回一个的是一个Wave_read类的实例，通过调用它的方法读取WAV文件的格式和数据。
    f = wave.open(wavFileName, "rb")
    # 读取格式信息
    # 一次性返回所有的WAV文件的格式信息，它返回的是一个组元(tuple)：声道数, 量化位数（byte单位）, 采
    # 样频率, 采样点数, 压缩类型, 压缩类型的描述。wave模块只支持非压缩的数据，因此可以忽略最后两个信息
    params = f.getparams()
    nchannels, sampwidth, framerate, nframes = params[:4]
    CutFrameNum = framerate * CutTimeDef / 1000
    print("CutFrameNum=%d" % (CutFrameNum))

    print("nchannels=%d" % (nchannels))
    print("sampwidth=%d" % (sampwidth))
    print("framerate=%d" % (framerate))
    print("nframes=%d" % (nframes))
    str_data = f.readframes(nframes)
    f.close()  # 将波形数据转换成数组
    # 需要根据声道数和量化单位，将读取的二进制数据转换为一个可以计算的数组
    wave_data = np.fromstring(str_data, dtype=np.short)
    wave_data.shape = -1, 2
    wave_data = wave_data.T
    temp_data = wave_data.T
    # StepNum = int(nframes/200)
    StepNum = int(CutFrameNum)
    StepTotalNum = 0
    stemp = 0
    while StepTotalNum < nframes:
        print("Stemp=%d" % (stemp))
        FileName = filedir + '/' + filename + '_' + str(stemp) + ext
        print("%s done!" %FileName)
        temp_dataTemp = temp_data[StepNum * (stemp):StepNum * (stemp + 1)]
        stemp = stemp + 1
        StepTotalNum = stemp * StepNum
        temp_dataTemp.shape = 1, -1
        temp_dataTemp = temp_dataTemp.astype(np.short)  # 打开WAV文档
        f = wave.open(FileName, "wb")  #
        # 配置声道数、量化位数和取样频率
        f.setnchannels(nchannels)
        f.setsampwidth(sampwidth)
        f.setframerate(framerate)
        # 将wav_data转换为二进制数据写入文件
        f.writeframes(temp_dataTemp.tostring())
        f.close()
    return stemp
# 导入文件
[filename, ext] = os.path.splitext(WavFile)
filedir = os.path.join('./',filename) + "_segmentation/"
WavNum = CutFile(WavFile)
```

### 对子音频频谱分析、时域分析
* 由于频域强度落在实数域上，应对其归一化；我采用了线性归一化方式，即：$ f = (f_0 - Min) / (Max - Min) $
* 在FFT中，由于取$N = f_s $，可推知第$X[k]$的第$k$个频谱线所对应实际频率$f = f_s/N *k $,使得数字频率与实际模拟频率相等

```python
# 进行0~1之间正则化，线性变化
def MaxMinNormalization(x,Max,Min):
    x = (x - Min) / (Max - Min)
    return x
# 进行数组归一化
def MMNormalization(original_array):
    abs_array = abs(original_array)
    max = np.max(abs_array)
    min = np.min(abs_array)
    if max != min:
        c = MaxMinNormalization(abs_array,max,min)
    else:
        c = np.zeros(original_array.shape)
    return c

# 绘制频谱图,其中d为展示频谱最大范围，默认值d取N/2
def GenerateFGraph(WavFile,d = -1):
    wave_data,nframes,framerate,width,channels = GenerateWavData(WavFile)
    # 采样点数，修改采样点数和起始位置进行不同位置和长度的音频波形分析
    N = framerate
    start = 0 #开始采样位置
    df = framerate/(N-1) # 分辨率
    freq = [df*n for n in range(0,N)] #N个元素
    wave_data2=wave_data[0][start:start+N]
    original_array=np.fft.fft(wave_data2)*2/N
    # 进行正则化，将赋值转变到0~1之间
    c = MMNormalization(original_array)
    #常规显示采样频率一半的频谱
    if d <= 0:
        d=int(len(c)/2)
    pylab.plot(freq[:d-1],abs(c[:d-1]),'r')
    pylab.show()
    
print("随机选取一个子文件，展示其波形图、频谱图")
n = random.randint(0,WavNum)
TestWavFile = filedir + filename + "_" + str(n) + ext
print("%s 号波形图"%str(n))
GenerateWavGraph(TestWavFile)
print("%s 号频谱图"%str(n))
GenerateFGraph(TestWavFile,d = 3000)
print("可以看出波形图不规律，频谱杂乱")
```

### 对于钢琴88音的中央C区进行波形分析、频谱分析
为了得到标准音的特征频率（基频），我也对钢琴88音进行了波形、频谱分析，并且期望滤除其谐波分量，记录基频，其方法如下：
+ step 1: 由于在得到频谱图时已将$X[k]$归一化，则可设定阈值$threshold$,先遍历搜索$X[k]>threshold$的特征频率,得到特征频率集合$\{f_k\}$
+ step 2: 再定义最远分辨距离$NBDistance$,对任意$i,j$,若$|f_i-f_j|≤NBDistance$，则去除$min\{X[f_i],X[f_j]\}$,即若两个特征频率相距过近，则应当去除幅度较小的特征频率；最终可留下最小特征频率集。

```python
# 对“秩”相近的特征频率幅值进行比较，去除幅度较小，且有相近特征频率的“无用”特征频率
def Simplify(f,a,distance_index):
    current = 0
    i = 0
    while i < f.size:
        if 0 < abs(f[i]-f[current]) <= distance_index:
            if(a[i]>a[current]):
                a = np.delete(a,current,axis = 0)
                f = np.delete(f,current,axis = 0)
                current = i - 1
            else:
                a = np.delete(a,i,axis = 0)
                f = np.delete(f,i,axis = 0)
            i = i - 1
        elif abs(f[i]-f[current]) > distance_index:
            current = i
        i = i+1
    return f,a

# 寻找相近特征频率谱线较大的值，最终返回最小特征频率集，及其幅度值
def FindFrequency(x):
    f_lst = []
    a_lst = []
    for i in range(int(x.shape[0]/2)):
        if x[i] > threshold:
            f_lst.append(i)
            a_lst.append(x[i])
    f_arr = np.array(f_lst)
    a_arr = np.array(a_lst)
    f_arr,a_arr = Simplify(f_arr,a_arr,distance_index = 5)
    return f_arr,a_arr

print("不妨选取最典型音阶，中央Do音")
TestWavFile = key_file + "061 " + "083"+ext
print("波形图")
GenerateWavGraph(TestWavFile)
print("频谱图")
GenerateFGraph(TestWavFile,d = 1000)
print("可以看出波形图较规律，且特征频率已经较为纯净")
print("")
print("下面对其进行再一步简化，争取得到唯一特征频率")

# 进行特征频率简化
wave_data,nframes,framerate,width,channels = GenerateWavData(TestWavFile)
# 采样点数，修改采样点数和起始位置进行不同位置和长度的音频波形分析
N = framerate
start = 0 #开始采样位置
wave_data2=wave_data[0][start:start+N]
original_array=np.fft.fft(wave_data2)*2/N
# 进行正则化，将赋值转变到0~1之间
c = MMNormalization(original_array)
f_arr,a_arr = FindFrequency(c)
print("Do FFT后的原始频谱X[k]\n{\n",original_array,"\n}")
print("Do特征频率：",f_arr)
print("Do特征频率对应归一化幅度",a_arr)
```

### 对分割好的测试wav子文件进行频谱测试，并根据音阶－频率表得出特征频谱
主要步骤如下:
+ step 1: 对切割好的wav子文件进行FFT,经线性归一化后得到其频谱数组
+ step 2: 对频谱数组进行简化操作，即根据已设定的$threshold$对得到的频率分量进行简化，得到其最小特征频率集$\{f_k\}min$(算法步骤见上面的CELL [对于钢琴88音的中央C区进行波形分析、频谱分析](### 对于钢琴88音的中央C区进行波形分析、频谱分析))
+ step 3: 利用音阶－频率表确定最小频率集$\{f_k\}min$对应的琴键序号，再利用宏定义中的KeyList得出其简谱符号<br> （其中，由于音阶－频谱表为增序序列，可利用二分查找法进行查找搜索） </br>

```python
# 二分查找，返回失败的位置（秩最大者）
def bin_search(data_set,val):
    #low 和high代表下标 最小下标，最大下标
    low = 0
    high = len(data_set)-1
    while low < high:# 只有当low小于High的时候证明中间有数
        mid = int((low+high) / 2 )
        if(val < data_set[mid]):
            high = mid
        else:
            low = mid + 1
    low = low - 1
    return low 

# 对音阶进行分类
def Classify(f,FDic):
    f_indexs = []
    for i in range(len(f)):
        index = bin_search(FDic,f[i])
        # 若音阶最低
        if index == -1:
            index = 0
        # 若音阶最高
        elif index == len(f)-1:
            pass
        else:
            gap = FDic[index + 1] - FDic[index]
            # 检查离目标频率最近的音阶
            if (f[i]-FDic[index]) > (gap/2):
                index = index + 1
        f_indexs.append(index)
    f_indexs_arr = np.array(f_indexs)
    return f_indexs_arr
```

```python
print("随机选取一个子文件，展示其波形图、频谱图")
n = random.randint(0,WavNum)
print("选取%d号子文件" %n)
TestWavFile = filedir + filename + "_" + str(n) + ext
# 进行特征频率简化
wave_data,nframes,framerate,width,channels = GenerateWavData(TestWavFile)
# 采样点数，修改采样点数和起始位置进行不同位置和长度的音频波形分析
N = framerate
start = 0 #开始采样位置
wave_data2=wave_data[0][start:start+N]
original_array=np.fft.fft(wave_data2)*2/N
# 进行正则化，将赋值转变到0~1之间
c = MMNormalization(original_array)
print(n,"号子文件　原始fft　Ｘ[k]\n{\n",original_array,"\n}")
f_arr,a_arr = FindFrequency(c)
print(n,"号子文件　经简化特征频率：",f_arr)
print(n,"号子文件　经简化特征频率对应归一化幅度",a_arr)
f_index_arr = Classify(f_arr,FDic)
print(n,"号子文件对应琴键序号：",f_index_arr)
print(n,"号子文件最终得到简谱符号：")
# 翻译为简谱
print("{",end = " ")
for j in range(len(f_index_arr)):
    print(KeyList[f_index_arr[j]],end = " ")
print("}",end = " ")
```

### 得到原始音乐的简谱
可重复以上操作，对每一个子文件逐一得出其简谱，最终将简谱相连可得到原始音乐的简谱，最终输出到文件（"<i>[filename]</i> ScoreText.txt"）中

```python
print("共有%d个子文件"%WavNum)
num = [str(n) for n in range(0,WavNum)]
with open(filename + "ScoreText.txt",'w') as f:
    for i in range(len(num)):
        WavFile = filedir + filename + "_" + num[i] + ext
        wave_data,nframes,framerate,width,channels = GenerateWavData(WavFile)
        # 采样点数，修改采样点数和起始位置进行不同位置和长度的音频波形分析
        N = framerate
        start = 0 #开始采样位置
        wave_data2=wave_data[0][start:start+N]
        original_array=np.fft.fft(wave_data2)*2/N
        # 进行正则化，将赋值转变到0~1之间
        c = MMNormalization(original_array)
        f_arr,a_arr = FindFrequency(c)
        f_index_arr = Classify(f_arr,FDic)
        # 翻译为简谱
        print("{",end = " ")
        print("{",end = " ",file = f)
        for j in range(len(f_index_arr)):
            print(KeyList[f_index_arr[j]],end = " ",file = f)
            print(KeyList[f_index_arr[j]],end = " ")
        print("}",end = " ",file = f)
        print("}",end = " ")
        # print(f_index_arr,end = '') 
        if (i+1) % 4 == 0:
            print("",file = f)
            print("")
```

## 总结
经过实验发现，对于大多数音乐，以上程序的吻合度并不高，其误差可能来自于以下方面:
+ 对频谱作了归一化处理，导致一些较弱的背景音乐对主声道的鉴别造成了干扰
+ 频谱未经过加窗处理，导致会有频率泄露等现象，造成频谱波动等
+ wav子文件的分割时长需要人工进行调整，否则输出的准确率大大降低
